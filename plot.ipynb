```python
import matplotlib.pyplot as plt
from math import sqrt
```

```python
sq = sqrt(3)
x = [1,3,9,27,81]
y = [4,9 + sq, 22 + 2*sq, 45 + 7*sq, 100 + 14* sq]

fig, ax = plt.subplots()

ax.plot(x,y,'ro-',linewidth = 2.0)

for x,y in zip(x,y):
    label = f'T({x})'
    plt.annotate(label,(x,y),textcoords="offset points",xytext= (0,10),ha = 'center')


fig.set_figwidth(10)
fig.set_figheight(8)
```

```python
def T_n(n):
    return (n ** math.log(2, 3)) * (8 + 2*math.sqrt(3)) + math.sqrt(n)*(-3 - 2*math.sqrt(3)) - 1


def G_x(n):
    return 9 * (n ** math.log(2, 3))
```

```python
fig, ax = plt.subplots()
X = np.arange(0,200)
T_ = list(map(T_n,X))
G_ = list(map(G_x,X))

ax.plot(X,T_,'r',label = 'T(n)')


x = [1,3,9,27,81]
y = [4,9 + sq, 22 + 2*sq, 45 + 7*sq, 100 + 14* sq]
ax.plot(x,y,'bo',linewidth = 2.0)

for x,y in zip(x,y):
    label = f'T({x})'
    plt.annotate(label,(x,y),textcoords="offset points",xytext= (0,10),ha = 'center')


ax.legend()
ax.grid()
fig.set_figwidth(10)
fig.set_figheight(8)
```

```python
def T_n(X):
    def T_n_(n):
        return (n ** math.log(2, 3)) * (8 + 2*math.sqrt(3)) + math.sqrt(n)*(-3 - 2*math.sqrt(3)) - 1
    return list(map(T_n_,X))
def G_x(X,c):
    def G_x_(n):
        return c * (n ** math.log(2, 3))
    return list(map(G_x_,X))
```

```python
fig, ax = plt.subplots()


X = [i for i in range(0,5000)]
T_ = T_n(X)
G_9 = G_x(X,9)
G_11 = G_x(X,11)

ax.plot(X,T_,'r',label = 'T(n)')

ax.plot(X,G_9,'b', label = '9 * G(n)')
ax.plot(X,G_11,'g', label = '11 * G(n)')
x_o = 1628
plt.annotate(f'x_o = {x_o}',(x_o,G_9[x_o]),textcoords="offset points",xytext= (10,0),ha = 'left',va= 'bottom')
ax.legend()
ax.grid()
fig.set_figwidth(10)
fig.set_figheight(8)
```
